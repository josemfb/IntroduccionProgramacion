# M칩dulos Propios

En `Python`, adem치s de poder crear nuestras propias funciones, podemos crear nuestros propios m칩dulos.

La estructura del c칩digo de un m칩dulo es la siguiente:

```python
## Variables del m칩dulo
PI = 3.1415925635
EULER = 2.71828182845

## Funciones del m칩dulo
def suma(a,b):
    return a + b

def resta(a,b):
    return a - b

## Sector "main" del archivo
if __name__ == "__main__":
    resultado = suma(2,3)
    print("El resultado es:", resultado)
```

En realidad son dos secciones principales en un m칩dulo: Lo que est치 en el `main` y lo que no.
    
Todo lo que **NO** est치 en el main, puede ser invocado al ser importado por otro archivo.

Todo lo que **SI** est치 en el main, se llama solamente al ejecutar el archivo directo con `Python`. Por ejemplo, escibiendo en la consola `python nombre_modulo.py`

Utilizando el c칩digo anterior y guard치ndolo en el archivo `modulo_matematico.py` podemos probar nuestro m칩dulo.

```python
from modulo_matematico import suma

print(suma(3,4))
```

```python
import modulo_matematico as mm

print(mm.resta(10,2))
print(mm.PI)
```

Que ocurrir칤a si agrego un `print("No estoy en el main")` al m칩dulo, en el sector fuera del `main`.

```python
from modulo_matematico import suma

print(suma(3,4))
```

```python
import modulo_matematico as mm

print(mm.resta(10,2))
print(mm.PI)
```

Cada vez que agrego c칩digo fuera del `main`, al ser importado nuestro m칩dulo, este se ejecuta.

Hay que tener cuidado con el c칩digo que agregamos fuera del `main`. Este deber칤a componerse solo de funciones y variables.

Crear nuestros propios m칩dulos nos permite tener el c칩digo m치s ordenado y agrupado.
